```python
import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from pathlib import Path

DATA = Path('C:/Users/Teo/Desktop')
```

```python
file = pd.read_csv(DATA / 'cadaver calibration tibia ITS data.csv')
file.head()
```

```python
matplotlib.rcParams['font.sans-serif'] = "Arial"
plt.rcParams['figure.figsize'] = [5.9*3.5,3.2*3.5]
plt.rcParams.update({'font.size': 20})
plt.rcParams.update({'font.weight': 'bold'})
fig, ax = plt.subplots(2,4)

BV_TV_m, BV_TV_b = np.polyfit(file.BV_TV_XCT2, file.BV_TV, 1)
BV_TV_r = (file['BV_TV'].corr(file['BV_TV_XCT2']))**2
ax[0][0].scatter(file.BV_TV_XCT2,file.BV_TV, color='black')
ax[0][0].set_xlabel("2G HR-pQCT", fontweight='bold')
ax[0][0].set_ylabel("\u03bCCT", fontweight='bold')
ax[0][0].set_title("BV/TV", fontweight='bold')
ax[0][0].plot(file.BV_TV_XCT2, (BV_TV_m*file.BV_TV_XCT2 + BV_TV_b), label = 'y = {:.2f}*x + {:.2f}\n $R^2$={:.2f}'.format(BV_TV_m, BV_TV_b, BV_TV_r), color='black')
ax[0][0].legend(loc='upper left', fontsize = 20, frameon=False)
ax[0][0].set_ylim(0.07,0.21)

pBV_TV_m, pBV_TV_b = np.polyfit(file.pBV_TV_XCT2, file.pBV_TV, 1)
pBV_TV_r = (file['pBV_TV'].corr(file['pBV_TV_XCT2']))**2
ax[0][1].scatter(file.pBV_TV_XCT2,file.pBV_TV, color='black')
ax[0][1].set_xlabel("2G HR-pQCT", fontweight='bold')
ax[0][1].set_ylabel("\u03bCCT", fontweight='bold')
ax[0][1].set_title("pBV/TV", fontweight='bold')
ax[0][1].plot(file.pBV_TV_XCT2, (pBV_TV_m*file.pBV_TV_XCT2 + pBV_TV_b), label = 'y = {:.2f}*x + {:.2f}\n $R^2$={:.2f}'.format(pBV_TV_m, pBV_TV_b, pBV_TV_r), color='black')
ax[0][1].legend(loc='upper left', frameon=False)
ax[0][1].set_ylim(0.04,0.19)

pTb_N_m, pTb_N_b = np.polyfit(file.pTb_N_XCT2, file.pTb_N, 1)
pTb_N_r = (file['pTb_N'].corr(file['pTb_N_XCT2']))**2
ax[0][2].scatter(file.pTb_N_XCT2,file.pTb_N, color='black')
ax[0][2].set_xlabel("2G HR-pQCT", fontweight='bold')
ax[0][2].set_ylabel("\u03bCCT", fontweight='bold')
ax[0][2].set_title("pTb.N", fontweight='bold')
ax[0][2].plot(file.pTb_N_XCT2, (pTb_N_m*file.pTb_N_XCT2 + pTb_N_b), label = 'y = {:.2f}*x + {:.2f}\n $R^2$={:.2f}'.format(pTb_N_m, pTb_N_b, pTb_N_r), color='black')
ax[0][2].legend(loc='upper left', frameon=False)
ax[0][2].set_ylim(1.3,1.9)

PR_Ratio_m, PR_Ratio_b = np.polyfit(file.PR_Ratio_XCT2, file.PR_Ratio, 1)
PR_Ratio_r = (file['PR_Ratio'].corr(file['PR_Ratio_XCT2']))**2
ax[1][0].scatter(file.PR_Ratio_XCT2,file.PR_Ratio, color='black')
ax[1][0].set_xlabel("2G HR-pQCT", fontweight='bold')
ax[1][0].set_ylabel("\u03bCCT", fontweight='bold')
ax[1][0].set_title("PR Ratio", fontweight='bold')
ax[1][0].plot(file.PR_Ratio_XCT2, (PR_Ratio_m*file.PR_Ratio_XCT2 + PR_Ratio_b), label = 'y = {:.2f}*x + {:.2f}\n $R^2$={:.2f}'.format(PR_Ratio_m, PR_Ratio_b, PR_Ratio_r), color='black')
ax[1][0].legend(loc='upper left', frameon=False)
ax[1][0].set_ylim(1,8)

pTb_Th_m, pTb_Th_b = np.polyfit(file.pTb_Th_XCT2, file.pTb_Th, 1)
pTb_Th_r = (file['pTb_Th'].corr(file['pTb_Th_XCT2']))**2
ax[0][3].scatter(file.pTb_Th_XCT2,file.pTb_Th, color='black')
ax[0][3].set_xlabel("2G HR-pQCT", fontweight='bold')
ax[0][3].set_ylabel("\u03bCCT", fontweight='bold')
ax[0][3].set_title("pTb.Th", fontweight='bold')
ax[0][3].plot(file.pTb_Th_XCT2, (pTb_N_m*file.pTb_Th_XCT2 + pTb_Th_b), label = 'y = {:.2f}*x + {:.2f}\n $R^2$={:.2f}'.format(pTb_Th_m, pTb_Th_b, pTb_Th_r), color='black')
ax[0][3].legend(loc='upper left', frameon=False)
ax[0][3].set_ylim(0.118,0.165)


rBV_TV_m, rBV_TV_b = np.polyfit(file.rBV_TV_XCT2, file.rBV_TV, 1)
rBV_TV_r = (file['rBV_TV'].corr(file['rBV_TV_XCT2']))**2
ax[1][1].scatter(file.rBV_TV_XCT2,file.rBV_TV, color='black')
ax[1][1].set_xlabel("2G HR-pQCT", fontweight='bold')
ax[1][1].set_ylabel("\u03bCCT", fontweight='bold')
ax[1][1].set_title("rBV/TV", fontweight='bold')
ax[1][1].plot(file.rBV_TV_XCT2, (rBV_TV_m*file.rBV_TV_XCT2 + rBV_TV_b), label = 'y = {:.2f}*x + {:.2f}\n $R^2$={:.2f}'.format(rBV_TV_m, rBV_TV_b, rBV_TV_r), color='black')
ax[1][1].legend(loc='upper left', frameon=False)
ax[1][1].set_ylim(0.015,0.05)

rTb_N_m, rTb_N_b = np.polyfit(file.rTb_N_XCT2, file.rTb_N, 1)
rTb_N_r = (file['rTb_N'].corr(file['rTb_N_XCT2']))**2
ax[1][2].scatter(file.rTb_N_XCT2,file.rTb_N, color='black')
ax[1][2].set_xlabel("2G HR-pQCT", fontweight='bold')
ax[1][2].set_ylabel("\u03bCCT", fontweight='bold')
ax[1][2].set_title("rTb.N", fontweight='bold')
ax[1][2].plot(file.rTb_N_XCT2, (rTb_N_m*file.rTb_N_XCT2 + rTb_N_b), label = 'y = {:.2f}*x + {:.2f}\n $R^2$={:.2f}'.format(rTb_N_m, rTb_N_b, rTb_N_r), color='black')
ax[1][2].legend(loc='upper left', frameon=False)
ax[1][2].set_ylim(1.1,1.7)

rTb_Th_m, rTb_Th_b = np.polyfit(file.rTb_Th_XCT2, file.rTb_Th, 1)
rTb_Th_r = (file['rTb_Th'].corr(file['rTb_Th_XCT2']))**2
ax[1][3].scatter(file.rTb_Th_XCT2,file.rTb_Th, color='black')
ax[1][3].set_xlabel("2G HR-pQCT", fontweight='bold')
ax[1][3].set_ylabel("\u03bCCT", fontweight='bold')
ax[1][3].set_title("rTb.Th", fontweight='bold')
ax[1][3].plot(file.rTb_Th_XCT2, (rTb_Th_m*file.rTb_Th_XCT2 + rTb_Th_b), label = 'y = {:.2f}*x + {:.2f}\n $R^2$={:.2f}'.format(rTb_Th_m, rTb_Th_b, rTb_Th_r), color='black')
ax[1][3].legend(loc='upper left', frameon=False)
ax[1][3].set_ylim(0.11,0.18)

# set the spacing between subplots
plt.subplots_adjust(left=0.05,
                    bottom=0.1, 
                    right=1.0, 
                    top=1.0, 
                    wspace=0.3, 
                    hspace=0.3)
plt.show()
```
