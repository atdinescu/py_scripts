```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

DATA = Path('F:\Scans_guineapig\ITS')
```

```python
ITS_lat = pd.read_csv(DATA / 'LAT'/ 'result.txt', sep = '\t', header = None)
ITS_lat.columns = ['Filename','BV_TV', 'pBV_TV', 'rBV_TV', 'aBV_TV', 'pBV_BV', 'rBV_BV', 'pTb.N', 'rTb.N', 'pTb.Th',
                   'rTb.Th', 'pTb.S', 'rTb.l', 'RR_junc', 'PR_junc', 'PP junc','NA']
#ITS_lat.drop(columns=['NA'])
```

```python
ITS_lat['PR_ratio'] = ITS_lat.apply(
    lambda row:(row.pBV_TV)/(row.rBV_TV), axis=1)
```

```python
ITS_lat['ID'] = ITS_lat['Filename'].astype(str).str.split('_').str[0]
ITS_lat['Location'] = ITS_lat['Filename'].astype(str).str.split('_').str[1]
ITS_lat['Subregion'] = ITS_lat['Filename'].astype(str).str.split('_').str[2]
ITS_lat.head()
```

```python
ITS_med = pd.read_csv(DATA / 'MED'/'result.txt', sep = '\t', header = None)
ITS_med.columns = ['Filename','BV_TV', 'pBV_TV', 'rBV_TV', 'aBV_TV', 'pBV_BV', 'rBV_BV', 'pTb.N', 'rTb.N', 'pTb.Th',
                   'rTb.Th', 'pTb.S', 'rTb.l', 'RR_junc', 'PR_junc', 'PP junc','NA']
```

```python
ITS_med['PR_ratio'] = ITS_med.apply(
    lambda row:(row.pBV_TV)/(row.rBV_TV), axis=1)
```

```python
ITS_med['ID'] = ITS_med['Filename'].astype(str).str.split('_').str[0]
ITS_med['Location'] = ITS_med['Filename'].astype(str).str.split('_').str[1]
ITS_med['Subregion'] = ITS_med['Filename'].astype(str).str.split('_').str[2]
ITS_med.head()
```

```python
ITS_merge = pd.concat([ITS_med,ITS_lat])
ITS_merge.shape
```

```python
GP_ID = pd.read_csv(DATA /'GP_fileID.txt', header = None, sep = '\t')
GP_ID.columns = ['GP','Month', 'ID']
GP_ID.head()
```

```python
ITS_merge = pd.merge(ITS_merge, GP_ID, left_on = 'ID', right_on = 'ID')
ITS_merge.head(10)
```

```python
ITS_BV_TV = ITS_merge.groupby(['Location','Subregion','Month'])['BV_TV'].mean().unstack(level=2)
ITS_BV_TV
```

```python
ITS_BV_TV_err = ITS_merge.groupby(['Location','Subregion','Month'])['BV_TV'].std().unstack(level=2)
ITS_BV_TV_err
```

```python
ITS_BV_TV.plot(kind='bar', yerr= ITS_BV_TV_err, capsize=3, title='BV/TV').get_figure().savefig('ITS_BV_TV.png', dpi=1200, bbox_inches="tight")
```

```python
ITS_PR_ratio = ITS_merge.groupby(['Location','Subregion','Month'])['PR_ratio'].mean().unstack(level=2)
ITS_PR_ratio
```

```python
ITS_PR_ratio_err = ITS_merge.groupby(['Location','Subregion','Month'])['PR_ratio'].std().unstack(level=2)
ITS_PR_ratio_err
```
