# Create one table with SBP results

```python
gp_data = pd.read_csv(DATA / 'result_gp_SBP_longitudinal.txt', sep = '\t', header = None)
gp_data.columns = ['Filename','Thickness', 'Porosity', 'BMD', 'TMD']
gp_data['ID'] = gp_data['Filename'].astype(str).str.split('_').str[0]
gp_data['Location'] = gp_data['Filename'].astype(str).str.split('_').str[1]
gp_data['Subregion'] = gp_data['Filename'].astype(str).str.split('_').str[2]
gp_data['Location_Subregion'] = gp_data['Location'] + '_' + gp_data['Subregion']
gp_data.head()
```

```python
gp_id = pd.read_csv(DATA /'GP_fileID.txt', header = None, sep = '\t')
gp_id.columns = ['GP','Month', 'ID','Location']
```

```python
gp_merge= pd.merge(gp_data, gp_id, left_on = 'ID', right_on = 'ID')
gp_merge.head()
```

```python
gp_merge.groupby(['Location_Subregion', 'Month']).mean()
```

```python
gp_merge.groupby(['Location_Subregion', 'Month']).std()
```

## Filter by individual GP

```python
def label_cohort (row):
    if ((row['GP'] == 'GP_10') or (row['GP'] == 'GP_11') or (row['GP'] == 'GP_12')):
        return 5
    elif ((row['GP'] == 'GP_9') or (row['GP'] == 'GP_13') or (row['GP'] == 'GP_14') or (row['GP'] == 'GP_15')):
        return 4
    elif ((row['GP'] == 'GP_8')  or (row['GP'] == 'GP_16') or (row['GP'] == 'GP_17')):
        return 3
    else:
        return 0
```

```python
gp_merge['Cohort'] = gp_merge.apply(lambda row: label_cohort(row), axis=1)
```

```python
gp_merge['GP_Location_Subregion'] = gp_merge['GP'] + '_' + gp_merge['Location_Subregion']
```

```python
gp_RT_MED = gp_merge[(gp_merge.Location_Subregion == "RT_MED")]
```

```python
gp_RT_MED.groupby(['GP', 'Month']).mean()
```

```python
gp_RT_MED.groupby(['Cohort', 'Month']).mean()
```

```python
import scipy.stats as stats

stats.f_oneway(gp_RT_MED['Thickness'][gp_RT_MED['Cohort'] == 3],
               gp_RT_MED['Thickness'][gp_RT_MED['Cohort'] == 4],
               gp_RT_MED['Thickness'][gp_RT_MED['Cohort'] == 5])
```

```python
stats.f_oneway(gp_RT_MED['Porosity'][gp_RT_MED['Cohort'] == 3],
               gp_RT_MED['Porosity'][gp_RT_MED['Cohort'] == 4],
               gp_RT_MED['Porosity'][gp_RT_MED['Cohort'] == 5])
```

# Statistical Analysis

```python
model = smf.mixedlm(("Thickness ~ Month"), gp_RT_MED, groups=gp_RT_MED["GP"]) #, reformula="~Month")
mdf = model.fit(method=["bfgs"])
print(mdf.summary())
```

```python
model = smf.mixedlm(("Porosity ~ Month"), gp_RT_MED, groups=gp_RT_MED["GP"]) #, reformula="~Month")
mdf = model.fit(method=["lbfgs"])
print(mdf.summary())
```

```python
ITS_merge["Month"] = ITS_merge["Month"].astype(str)
#ITS_merge_5months["Month"] = ITS_merge_5months["Month"].astype(str)

hue_order = list(['2', '3', '4', '5'])
order = ["RT_LAT","RT_MED"]
#order = ["RF_LAT", "RF_MED", "RT_LAT","RT_MED"]
#order = ["LF_LAT", "LF_MED", "LT_LAT","LT_MED","RF_LAT", "RF_MED", "RT_LAT","RT_MED"]


hue_plot_params = {
    'data': ITS_merge,
    'x': 'Location_Subregion',
    "order": order,
    "hue": "Month",
    "hue_order": hue_order
}
```

```python
import seaborn as sns

def ITS_plot(ITS_param, ITS_param_units):
    sns.set_context(rc = {'patch.linewidth': 1.0})
    ax = sns.barplot(**hue_plot_params, y=ITS_param, estimator=np.mean, ci=68, capsize=.15, errcolor='black') 
    #,fill=False)
    ax = sns.stripplot(**hue_plot_params, y=ITS_param, color='black', size=14, dodge=True)
    
    #ax = sns.barplot(**hue_plot_params, y=ITS_param, estimator=np.mean, ci=68, capsize=.2, errcolor='black') #,fill=False)
    #ax = sns.swarmplot(**hue_plot_params, y=ITS_param, color='black', dodge=True)
    
    ax.set(xlabel = 'Location')
    ax.set(ylabel = ITS_param_units)
    #ax.set_title(ITS_param, fontdict={'fontsize': 44, 'weight': 'bold'})

    ax.legend_.remove()
    #ax.legend(loc='upper center',ncol=4)
    ax.spines['top'].set_visible(True)
    ax.spines['right'].set_visible(True)

    annotator = Annotator(ax, pairs, **hue_plot_params, y=ITS_param,show_non_significant=False)
    annotator.configure(test='t-test_ind', comparisons_correction='BH', correction_format="replace", text_format='star', color='black', loc='inside',pvalue_thresholds=[[1e-4, "*"], [1e-3, "*"], [1e-2, "*"], [0.05, "*"],[1, ""]])
    annotator.apply_and_annotate()
 
    img_name = str(DATA) + "/ITS_" + ITS_param + ".png"
    plt.tight_layout()

    plt.savefig(img_name, bbox_inches = "tight")
```

```python
#def ITS_subplots(ITS_param1, ITS_param2, ITS_param3, ITS_param4):
def ITS_subplots(ITS_param1, ITS_param_units1, ITS_param2, ITS_param_units2, ITS_param3, ITS_param_units3, ITS_param4, ITS_param_units4):
    plt.subplot(2,2,1)
    ITS_plot(ITS_param1, ITS_param_units1)
    plt.xticks(ha='center')
    plt.subplot(2,2,2)
    ITS_plot(ITS_param2, ITS_param_units2)
    plt.xticks(ha='center')
    plt.subplot(2,2,3)
    ITS_plot(ITS_param3, ITS_param_units3)
    plt.xticks(ha='center')
    plt.subplot(2,2,4)
    ITS_plot(ITS_param4,ITS_param_units4)
    plt.xticks(ha='center')
    
    img_name = str(DATA) + "/ITS_subplots.png"
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
    plt.savefig(img_name) #, bbox_inches = "tight")	
```
