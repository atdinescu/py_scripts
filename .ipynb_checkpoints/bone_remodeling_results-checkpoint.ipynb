```python
import numpy as np
import matplotlib.pyplot as plt

fd = open('../E0015896_ITS_L_1.raw', 'rb')
fd_sed = open('../E0015896_ITS_L_1.sed','rb')
dimX = 124
dimY = 166
dimZ = 74
f = np.fromfile(fd, dtype=np.uint8,count=dimX*dimY*dimZ)
f_sed = np.fromfile(fd_sed, dtype=np.float64,count=dimX*dimY*dimZ)
im = f.reshape((dimZ, dimY, dimX))
im_sed = f_sed.reshape((dimZ, dimY, dimX))
fd.close()
fd_sed.close()

im2d = im[0,:,:]
im2d_sed = im_sed[0,:,:]

print(im2d[:,:])
print(im2d_sed[:,:])
#plt.imshow(im)
#plt.savefig("E0015896_ITS_L_8.png")

# plot slice of bone image
plt.rcParams['figure.figsize'] = [20, 20]
plt.imshow(im2d[:,:])
```

```python
# plot colormap of SED value for slice above
fig, ax = plt.subplots()
plt.rcParams['figure.figsize'] = [20, 20]

ax.matshow(im2d, cmap=plt.cm.Blues)

for i in range(dimX):
    for j in range(dimY):
        c = im2d[j,i]
        ax.text(i,j,str(c),va='center',ha='center')
```
